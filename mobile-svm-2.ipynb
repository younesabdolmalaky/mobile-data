```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.csv')
df
```

```python
df.isnull().sum()
```

```python
X = df.loc[:, df.columns != "price_range"].to_numpy()
Y = df['price_range'].to_numpy()
```

```python
from sklearn.svm import SVC
from sklearn import metrics
svm_classifier = SVC(C=1 ,kernel = 'linear',probability = True)
svm_classifier.fit(X[:1500], Y[:1500])
p = svm_classifier.predict(X[1500:])
e = Y[1500:]
print(metrics.classification_report(e, p))
```

```python
df_test = pd.read_csv('test.csv')
df_test.drop(['id'] ,axis=1,inplace =True)
df_test
```

```python
probability = svm_classifier.predict_proba(df_test.to_numpy())
result = svm_classifier.predict(df_test.to_numpy())
```

```python
temp_df = pd.DataFrame(probability)
probability = temp_df.max(axis = 1).to_numpy()
```

```python
new_data_frame = df_test
```

```python
new_data_frame['result'] = pd.Series(result)
new_data_frame['probability'] = pd.Series(probability)
new_data_frame
```

```python
new_data_frame.to_csv('mobile-svm-2.csv')
```
