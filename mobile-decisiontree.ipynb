```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import tree
```

```python
df = pd.read_csv('train.csv')
df
```

```python
X = df.iloc[ : , df.columns != 'price_range'].to_numpy()
Y = df['price_range'].to_numpy()
```

```python
from sklearn.tree import DecisionTreeClassifier
des_tree_classifier = DecisionTreeClassifier(criterion= 'entropy')
```

```python
des_tree_classifier.fit(X[:1500],Y[:1500])
```

```python
from sklearn import metrics
```

```python
p = des_tree_classifier.predict(X[1500:])
e = Y[1500:]
print(metrics.classification_report(e,p))
```
